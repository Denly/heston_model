```python
import pandas as pd
from numpy import sqrt,mean,log,diff

# the data is here 
# https://docs.google.com/spreadsheets/d/1vN9MPBOyoh1Kf4OC0GlCF4kKdgmXholU8YUyRQjPBaE/edit?usp=sharing
# https://www.taifex.com.tw/cht/3/optDailyMarketView

df = pd.read_csv("txo 2021.05.01_2021.05.31.csv",  engine='python')
df.columns = ['symbol', 'exp', 'strik', 'put_call', 'o', 'h', 'l', 'c', 'v', 'settlement', '_','_','_','_','_','_', 'day_night','_']
df
```

```python
# replace mandarin to english
df = df.replace(to_replace ="�R�v", value ="call")\
        .replace(to_replace ="���v", value ="put")\
        .replace(to_replace ="�@��", value ="day")\
        .replace(to_replace ="�L��", value ="night")

df
```

```python
df = df[~df['day_night'].isin(['night'])] # drop the row is night
df = df[~df['settlement'].isin(['0'])] # drop the row that settlement is zero (usually strik price too far)
df['settlement'] = df['settlement'].astype(float)

close = df['2021/05/03':'2021/05/31']['settlement']
df
```

```python
# use the daily data of Google(NASDAQ: GOOG) from 01/2016 to 08/2016 
#close = df['2021/05/03':'2021/05/31']['c']

r = diff(log(close))
r_mean = mean(r)
diff_square = [(r[i]-r_mean)**2 for i in range(0,len(r))]
std = sqrt(sum(diff_square)*(1.0/(len(r)-1)))
vol = std*sqrt(252)
vol
```
