```python
#爬蟲函數 
def craw_new_data(year,month,date):
    import requests
    import pandas as pd
    from bs4 import BeautifulSoup
    import numpy
    import csv
    import time

            
    info = f'{year}/{month}/{date}'
    url = 'https://www.taifex.com.tw/cht/3/optDailyMarketReport'
    payload = {'queryType':'2',
               'marketCode':'0',
               'dateaddcnt':'',
               'commodity_id':'TXO',
               'commodity_id2':'',
               'MarketCode':'0',
               'commodity_idt':'TXO',
               'commodity_id2t':'',
               'commodity_id2t2':'',
               'queryDate':info
              }
    encoding = 'utf8'
    r = requests.post(url,data=payload)
    r.encoding = encoding
    soup = BeautifulSoup(r.content, 'html.parser')

    data = pd.DataFrame(columns=['契約','到期月份(週別)','履約價','買賣權','開盤價','最高價','最低價','最後成交價','結算價','盤後交易時段成交量','一般交易時段成交量','合計成交量','未沖銷契約量'])
    append_list = {}
    count = 0
    for i in soup.find_all('tr'):
        for j in i.find_all('td',class_ = '12bk'):
            count+=1
            if count > 18:
                if (count-18)%13 == 0:
                    data = data.append(append_list, ignore_index=True)
                    append_list = {}
                    append_list['契約'] = j.text.replace('\n','').replace(' ','').replace('\t','').replace('\r','')
                elif (count-18)%13 == 1:
                    append_list['到期月份(週別)'] = j.text.replace('\n','').replace(' ','').replace('\t','').replace('\r','')
                elif (count-18)%13 == 2:
                    append_list['履約價'] = j.text.replace('\n','').replace(' ','').replace('\t','').replace('\r','')
                elif (count-18)%13 == 3:
                    append_list['買賣權'] = j.text.replace('\n','').replace(' ','').replace('\t','').replace('\r','')
                elif (count-18)%13 == 4:
                    append_list['開盤價'] = j.text.replace('\n','').replace(' ','').replace('\t','').replace('\r','')
                elif (count-18)%13 == 5:
                    append_list['最高價'] = j.text.replace('\n','').replace(' ','').replace('\t','').replace('\r','')
                elif (count-18)%13 == 6:
                    append_list['最低價'] = j.text.replace('\n','').replace(' ','').replace('\t','').replace('\r','')
                elif (count-18)%13 == 7:
                    append_list['最後成交價'] = j.text.replace('\n','').replace(' ','').replace('\t','').replace('\r','')
                elif (count-18)%13 == 8:
                    append_list['結算價'] = j.text.replace('\n','').replace(' ','').replace('\t','').replace('\r','')
                elif (count-18)%13 == 9:
                    append_list['盤後交易時段成交量'] = j.text.replace('\n','').replace(' ','').replace('\t','').replace('\r','')
                elif (count-18)%13 == 10:
                    append_list['一般交易時段成交量'] = j.text.replace('\n','').replace(' ','').replace('\t','').replace('\r','')
                elif (count-18)%13 == 11:
                    append_list['合計成交量'] = j.text.replace('\n','').replace(' ','').replace('\t','').replace('\r','')
                elif (count-18)%13 == 12:
                    append_list['未沖銷契約量'] = j.text.replace('\n','').replace(' ','').replace('\t','').replace('\r','')

    data.to_csv(f'option_data_{year}_{month}_{date}.csv',index=False,encoding='cp950')
```

```python
#選周結算的契約,距結算剩餘日


def settle_select(year,month,date):

    import pandas as pd
    import numpy
    import time
    import datetime
    
    d1 = datetime.datetime(year,month,date)
    
    if d1.weekday()==2: #周三
        left_days=7
        job_days=5
        if month<10:
            month1= '0'+ str(month)
            month2= '0'+ str(month+1)
        else:
            month1= str(month)
            month2= str(month+1)

        if date>14 and date<22 :
            text1= str(year) + month1 +'W4'
            return(text1,left_days,job_days)
        elif date>7 and date<15 :  
            text1= str(year) + month1 
            return(text1,left_days,job_days)
        elif date>0 and date<8 :  
            text1= str(year) + month1  +'W2'
            return(text1,left_days,job_days)

        elif date>=22 :
            if date>=25 :
                if month<12:
                    text1= str(year) + month2  +'W1'
                    return(text1,left_days,job_days)
                else:
                    text1= str(year+1) + '01'  +'W1'
                    return(text1,left_days,job_days)
            elif date==24 and (month==2 or month==4  or month==6  or month==9 or month==11) :
                text1= str(year) + month2  +'W1'
                return(text1,left_days,job_days)

            else:
                text1= str(year) + month1  +'W5'
                return(text1,left_days,job_days)
            
    elif d1.weekday()==3:
        left_days=6
        job_days=4
        if month<10:
            month1= '0'+ str(month)
            month2= '0'+ str(month+1)
        else:
            month1= str(month)
            month2= str(month+1)

        if date>15 and date<23 :
            text1= str(year) + month1 +'W4'
            return(text1,left_days,job_days)
        elif date>8 and date<16 :  
            text1= str(year) + month1 
            return(text1,left_days,job_days)
        elif date>1 and date<9 :  
            text1= str(year) + month1  +'W2'
            return(text1,left_days,job_days)
        
        elif date<=1 :  
            text1= str(year) + month1  +'W1'
            return(text1,left_days,job_days)
        
        
        elif date>=23 :
            if date>=26 :
                if month<12:
                    text1= str(year) + month2  +'W1'
                    return(text1,left_days,job_days)
                else:
                    text1= str(year+1) + '01'  +'W1'
                    return(text1,left_days,job_days)
            elif date==25 and (month==2 or month==4  or month==6  or month==9 or month==11) :
                text1= str(year) + month2  +'W1'
                return(text1,left_days,job_days)

            else:
                text1= str(year) + month1  +'W5'
                return(text1,left_days,job_days)
        
    elif d1.weekday()==4:
        left_days=5
        job_days=3
        if month<10:
            month1= '0'+ str(month)
            month2= '0'+ str(month+1)
        else:
            month1= str(month)
            month2= str(month+1)

        if date>16 and date<24 :
            text1= str(year) + month1 +'W4'
            return(text1,left_days,job_days)
        elif date>9 and date<17 :  
            text1= str(year) + month1 
            return(text1,left_days,job_days)
        elif date>2 and date<10 :  
            text1= str(year) + month1  +'W2'
            return(text1,left_days,job_days)
        elif date<=2 :  
            text1= str(year) + month1  +'W1'
            return(text1,left_days,job_days)
        
        elif date>=24 :
            if date>=27 :
                if month<12:
                    text1= str(year) + month2  +'W1'
                    return(text1,left_days,job_days)
                else:
                    text1= str(year+1) + '01'  +'W1'
                    return(text1,left_days,job_days)
            elif date==26 and (month==2 or month==4  or month==6  or month==9 or month==11) :
                text1= str(year) + month2  +'W1'
                return(text1,left_days,job_days)

            else:
                text1= str(year) + month1  +'W5'
                return(text1,left_days,job_days)

            
        
    elif d1.weekday()==0:
        left_days=2
        job_days=2
        if month<10:
            month1= '0'+ str(month)
            month2= '0'+ str(month+1)
        else:
            month1= str(month)
            month2= str(month+1)

        if date>19 and date<27 :
            text1= str(year) + month1 +'W4'
            return(text1,left_days,job_days)
        elif date>12 and date<20 :  
            text1= str(year) + month1 
            return(text1,left_days,job_days)
        elif date>5 and date<13 :  
            text1= str(year) + month1  +'W2'
            return(text1,left_days,job_days)
        elif date<=5 :  
            text1= str(year) + month1  +'W1'
            return(text1,left_days,job_days)
        
        elif date>=27  :
            if date>=30 :
                if month<12:
                    text1= str(year) + month2  +'W1'
                    return(text1,left_days,job_days)
                else:
                    text1= str(year+1) + '01'  +'W1'
                    return(text1,left_days,job_days)
            elif date==29 and (month==2 or month==4  or month==6  or month==9 or month==11) :
                text1= str(year) + month2  +'W1'
                return(text1,left_days,job_days)

            else:
                text1= str(year) + month1  +'W5'
                return(text1,left_days,job_days)
        
        
    elif d1.weekday()==1:
        left_days=1
        job_days=1
        if month<10:
            month1= '0'+ str(month)
            month2= '0'+ str(month+1)
        else:
            month1= str(month)
            month2= str(month+1)

        if date>20 and date<28 :
            text1= str(year) + month1 +'W4'
            return(text1,left_days,job_days)
        elif date>13 and date<21 :  
            text1= str(year) + month1 
            return(text1,left_days,job_days)
        elif date>6 and date<14 :  
            text1= str(year) + month1  +'W2'
            return(text1,left_days,job_days)
        elif date<=6 :  
            text1= str(year) + month1  +'W1'
            return(text1,left_days,job_days)
        
        elif date>=28  :
            if date>=31 :
                if month<12:
                    text1= str(year) + month2  +'W1'
                    return(text1,left_days,job_days)
                else:
                    text1= str(year+1) + '01'  +'W1'
                    return(text1,left_days,job_days)
            elif date==30 and (month==2 or month==4  or month==6  or month==9 or month==11) :
                text1= str(year) + month2  +'W1'
                return(text1,left_days,job_days)

            else:
                text1= str(year) + month1  +'W5'
                return(text1,left_days,job_days)
        
```

```python
#主程式

from QuantLib import *
import datetime
import pandas as pd

year1=2021
month1=6
date1=25

settle_contract,left_days,job_days = settle_select(year1,month1,date1) # 算出哪個周契約 剩餘幾天結算

calendar = Taiwan()
calculation_date = Date(date1, month1, year1)         # 要找的日期
expiration_dates = [ Date(date1, month1, year1)+left_days ]    # 周結算日
day_count = Actual365Fixed()
Settings.instance().evaluationDate = calculation_date

              
craw_new_data(year1,month1,date1)
text00 = 'option_data_' + str(year1) + "_" +str(month1) +"_"+str(date1)+".csv"   #存檔檔名

read_option_ver0 = pd.read_csv(text00, encoding= 'big5')

read_option_ver0['到期月份(週別)'] = read_option_ver0['到期月份(週別)'].astype('str') 

choice1 = read_option_ver0[read_option_ver0['到期月份(週別)'] == settle_contract ]  #只選要的契約

choice1.reset_index(inplace=True)
choice1.drop(columns=["index"],inplace=True)
choice2=choice1.copy()


if choice2.shape[0]>0:
    spot_price = float(choice2.at[0, '履約價'])+float(choice2.at[0, '結算價'])-float(choice2.at[1, '結算價'])  #算標的物價

for i in range(0,choice2.shape[0]):
    choice2.at[i, 'moneyness'] = float(choice2.at[i, '履約價'])/spot_price   #標出每個履約價距離價平多遠

choice3 = choice2[choice2['moneyness'] >= 0.95 ]    #只選擇價平附近0.95~1.05的契約
choice4 = choice3[choice3['moneyness'] <= 1.05 ]
 
choice4.reset_index(inplace=True)
choice4.drop(columns=["index"],inplace=True)
choice5=choice4.copy()

spot_temp = 0 
count_temp = 0

choice5['結算價'] = choice5['結算價'].astype(float)
choice5['履約價'] = choice5['履約價'].astype(float)


exercise = EuropeanExercise( Date(date1, month1, year1)+left_days)   

S = QuoteHandle(SimpleQuote(spot_price))  
r = YieldTermStructureHandle(FlatForward(0, TARGET(), 0.01, Actual365Fixed())) # Rf=1%
q1 = YieldTermStructureHandle(FlatForward(0, TARGET(), 0, Actual365Fixed())) # d = 0%
                
sigma = BlackVolTermStructureHandle(BlackConstantVol(0, TARGET(), 0.20, Actual365Fixed())) # 初始值
process = BlackScholesMertonProcess(S,q1,r,sigma) 

for i in range(0,choice5.shape[0],2):
    call_price_temp = choice5.at[i+1, '結算價'] +  spot_price - choice5.at[i, '履約價']
    payoff = PlainVanillaPayoff(Option.Call,(float(choice5.at[i, '履約價'])))
    option = EuropeanOption(payoff,exercise)
    try:
        choice5.at[i, 'vol']=(option.impliedVolatility(choice5.at[i, '結算價'], process))  # 計算隱含波動率
    except:
        continue


strike_list = choice5[choice5['買賣權'] == 'Call' ] ["履約價"].values.tolist()
vol_list = choice5[choice5['買賣權'] == 'Call' ] ["vol"].values.tolist()
data = [vol_list]

implied_vols = Matrix(len(strike_list), len(expiration_dates))
for i in range(implied_vols.rows()):
    for j in range(implied_vols.columns()):
        implied_vols[i][j] = data[j][i]
        
black_var_surface = BlackVarianceSurface(
    calculation_date, calendar, 
    expiration_dates, strike_list, 
    implied_vols, day_count)

v0 = 0.00005; kappa = 0.2; theta = 0.03; rho = -0.1; sigma = 10;  # 初始值

process = HestonProcess(r, q1, 
                           QuoteHandle(SimpleQuote(spot_price)), 
                           v0, kappa, theta, sigma, rho)
model = HestonModel(process)
engine = AnalyticHestonEngine(model)   

heston_helpers = []
black_var_surface.setInterpolation("bicubic")
one_year_idx =0 
date = expiration_dates[one_year_idx]
for j, s in enumerate(strike_list):     
    t = job_days  
    p = Period(t, Days)
    sigma = data[one_year_idx][j]
    helper = HestonModelHelper(p, calendar, spot_price, s, 
                                  QuoteHandle(SimpleQuote(sigma)),
                                  r, 
                                  q1)
    helper.setPricingEngine(engine)
    heston_helpers.append(helper)
lm = LevenbergMarquardt(1e-8, 1e-8, 1e-8)
model.calibrate(heston_helpers, lm, EndCriteria(500, 50, 1.0e-8,1.0e-8, 1.0e-8))    #跑heston
theta, kappa, sigma, rho, v0 = model.params()
avg = 0.0

for i, opt in enumerate(heston_helpers):  #重新整理數據

        if choice5.at[2*i, '履約價']<=spot_price:
            choice5.at[2*i, 'Model Value']=opt.marketValue()+spot_price-choice5.at[2*i, '履約價']
        else:
            choice5.at[2*i, 'Model Value']=opt.marketValue()
            
        if choice5.at[2*i+1, '履約價']<=spot_price:
            choice5.at[2*i+1, 'Model Value']=opt.marketValue()
        else:
            choice5.at[2*i+1, 'Model Value']=choice5.at[2*i, '履約價']+opt.marketValue()-spot_price
        
        choice5.at[2*i, 'Error'] = choice5.at[2*i, '結算價'] - choice5.at[2*i, 'Model Value']
        choice5.at[2*i+1, 'Error'] = choice5.at[2*i+1, '結算價'] - choice5.at[2*i+1, 'Model Value']
            
data1 = choice5[choice5['買賣權'] == 'Call'].round(2)  
data1.drop(columns=["契約", "開盤價", "最高價", "最低價", "最後成交價", "盤後交易時段成交量", "一般交易時段成交量",
                        "合計成交量", "未沖銷契約量", "moneyness", "買賣權"],inplace=True)
data1.reset_index(inplace=True)
data1.drop(columns=["index"],inplace=True)
    
data2 = choice5[choice5['買賣權'] == 'Put'].round(2)  
data2.drop(columns=["契約", "開盤價", "最高價", "最低價", "最後成交價", "盤後交易時段成交量", "一般交易時段成交量",
                        "合計成交量", "未沖銷契約量", "moneyness", "買賣權"],inplace=True)
data2.reset_index(inplace=True)
data2.drop(columns=["index"],inplace=True)
    
data1['Put 市場價']=data2['結算價']
data1['Put Heston價']=data2['Model Value']
data1['Put模型誤差']=data2['Error']
    
data1 = data1[['到期月份(週別)','Error','Model Value','結算價','履約價','Put 市場價','Put Heston價','Put模型誤差','vol']]
data1.columns=['到期月份(週別)','Call模型誤差','Call Heston價','Call 市場價','履約價',
                   'Put 市場價','Put Heston價','Put模型誤差','隱含波動率']

text01 = text00[0:-4] + '_heston.csv' 
data1.to_csv(text01,index=False,encoding='cp950')



print("資料時間為:",year1,"年",month1,"月",date1,"日之收盤價        ", end='')
print("距離契約到期日剩下:",left_days,"天")
print("模型參數：", end='')
print( "theta = %f, kappa = %f, sigma = %f, rho = %f, v0 = %f" % (theta, kappa, sigma, rho, v0))
print(data1)
```
